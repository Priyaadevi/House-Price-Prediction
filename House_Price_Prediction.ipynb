```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], kde=True, bins=100)
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation with SalePrice(Default Matrix)
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Correlation with SalePrice
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
correlation_with_price = corr['price'].sort_values(ascending=False)
print("Top 10 positively correlated features:\n", correlation_with_price.head(10))
print("\nTop 10 negatively correlated features:\n", correlation_with_price.tail(10))
```

```python
features = ['price','sqft_living', 'sqft_above','bathrooms']

for feature in features:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'Sale Price vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```
